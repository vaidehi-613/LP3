```python
df.dropna(inplace=True)
```

```python
df.drop(['Email No.'],axis=1, inplace=True)
X=df.drop(['Prediction'],axis=1)
Y=df['Prediction']
```

```python
df.head()
```

```python
from sklearn.preprocessing import scale
X=scale(X)
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.3, random_state=42)
```

```python
Status = []
'''for i in range(0,df.shape[0]):
    if df['Prediction'][i]==0:
        Status.append("Not spam")
    else:
        Status.append("Spam")'''

Status = ["Not spam" if df['Prediction'][i]==0 else "Spam" for i in range(0,df.shape[0])]
        
df['Status'] = Status
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train,Y_train)
y_predict=knn.predict(X_test)
```

```python
print('Prediction',y_predict)
```

```python
print('KNN Accuracy:', metrics.accuracy_score(Y_test,y_predict))
```

```python
from sklearn.metrics import confusion_matrix
print('KNN Confusion Matrix:', metrics.confusion_matrix(Y_test,y_predict))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_predict))
```

```python
model=SVC(C=1)
model.fit(X_train, Y_train)
y_pred=model.predict(X_test)
```

```python
metrics.confusion_matrix(y_true=Y_test, y_pred=y_pred)
```

```python
print("SVM accuracy = ",metrics.accuracy_score(Y_test,y_pred))
```

```python
print('Classification Report:', metrics.classification_report(Y_test, y_pred))
```
